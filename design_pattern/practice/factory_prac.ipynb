```python
from abc import ABC, abstractmethod

class IChair(ABC):
    
    @staticmethod
    @abstractmethod
    def get_dimensions():
        "An abstract interface method"
        
class SmallChair(IChair):
    _height :int = 40
    _width :int = 40
    _depth :int = 40
    
    def get_dimensions(self) -> dict:
        return {"height":self._height,
                "_width":self._width,
                "_depth":self._depth}
        
class MediumChair(IChair):
    _height :int = 60
    _width :int = 60
    _depth :int = 60
    
    def get_dimensions(self) -> dict:
        return {"height":self._height,
                "_width":self._width,
                "_depth":self._depth}
        
class LargeChair(IChair):
    _height :int = 80
    _width :int = 80
    _depth :int = 80
    
    def get_dimensions(self) -> dict:
        return {"height":self._height,
                "_width":self._width,
                "_depth":self._depth}
    
class ChairFactory:
    CHOICE = {"LC":LargeChair(),
              "MC":MediumChair(),
              "SC":SmallChair()}
    
    @staticmethod
    def get_chair(chair :str) -> IChair:
        if chair in ChairFactory.CHOICE:
            return ChairFactory.CHOICE[chair]


CHAIR = ChairFactory.get_chair("SC")
print(CHAIR.get_dimensions())
    
```

```python
from abc import ABC, abstractmethod

class IFurnitureFactory(ABC):
    "Interface for Funiture Factory"
    @staticmethod
    @abstractmethod
    def get_furniture(type):
        "Interface of get furniture factory"
        

class FurnitureFactory(IFurnitureFactory):
    @staticmethod
    def get_furniture(type):
        try: #chair(s,m,l), table(s,m,l)
            if type in ["sc","mc","lc"]:
                return ChairFactory().get_chair(type[0])
            if type in ["st","mt","lt"]:
                return TableFactory().get_table(type[0])
            raise Exception("Not Factory which can serve your request exists")
        except Exception as e:
            print(e)
        return None
            


    
```

```python
CHAIR = FurnitureFactory.get_furniture("sc")
TABLE = FurnitureFactory.get_furniture("lt")
```

```python
print(CHAIR.get_dimensions())
print(TABLE.get_dimensions())
```
